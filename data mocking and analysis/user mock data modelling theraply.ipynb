## User Data Mocking: Predicting the Waiting Time to See a Therapist

### How long do people have to wait until they can get the first appointment with a therapist on Theraply app? We are going to create a mock data of a user named 'Mr. A' in Berlin, who speaks only English, is having anxiety symptomps, and looking for a therapist who accepts GKV/public insurance payment.

```python
import pandas as pd
import random
```

```python
df = pd.read_csv('/Users/tikaramadhini/Documents/theraplydata/thedata/final list of therapists and other info.csv', encoding_errors='ignore', sep=';', index_col='Number')
```

```python
df.head(5)
```

### Filtering search options based on needs

```python
mr_a = (df['Language'].str.contains('English') & df['Insurance'].str.contains('Public') & df['Expertise'].str.contains('Anxiety'))
```

```python
dfmr_a = df.loc[mr_a, ['Name','WaitingTime(Weeks)']]
dfmr_a
```

```python
#dropping therapists that are currently unavailable
dfmr_a.drop(dfmr_a[dfmr_a['WaitingTime(Weeks)'].str.contains('not available for new patients')].index, inplace=True)
dfmr_a
```

### Let's create a mock contacted therapists data by randomly selecting 10 therapists Mr. A would try to contact out of the available therapists list.

```python
selectedtherapists = dfmr_a.sample(n=10)
selectedtherapists
```

```python
selectedtherapists.to_csv('mock_selectedtherapists.csv', index=False)
```
